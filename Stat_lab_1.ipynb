```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

true_mean = 10
sigma = 0.2
n = 100

sample = np.random.normal(true_mean, sigma, n)
mean = np.mean(sample)
se = sigma / np.sqrt(n)

ci_low = mean - 1.96 * se
ci_high = mean + 1.96 * se

print("Sample mean:", mean)
print("95% CI:", ci_low, ci_high)
```
